```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Rename columns
df = df.rename(columns={"AVERAGE of 3.2.B.Environment3": "Table 3 (3.2)"})
df = df.rename(columns={"AVERAGE of 5.6.B.Environment3": "Table 5 (5.6)"})
df = df.rename(columns={"AVERAGE of 9.5.B.Environment3": "Table 9 (9.5)"})
df = df.rename(columns={"AVERAGE of Greenhouse Outside Ambient3": "Outside"})

# Melt to long format
df_long = df.melt(id_vars=["Timestamp", "year"], var_name="variable", value_name="value")

# Define palette for consistency
variables = df_long["variable"].unique()
palette = dict(zip(variables, sns.color_palette("tab10", n_colors=len(variables))))

# Create figure with two axes: main chart + histogram
fig, (ax_main, ax_hist) = plt.subplots(
    1, 2, figsize=(10,5), gridspec_kw={"width_ratios":[3,1]}, sharey=True
)

# --- Main scatter + line ---
sns.scatterplot(data=df_long, x="Timestamp", y="value",
                hue="variable", palette=palette, marker=".", ax=ax_main)
sns.lineplot(data=df_long, x="Timestamp", y="value",
             hue="variable", palette=palette, alpha=0.1, legend=False, ax=ax_main)

ax_main.set_title("Greenhouse: Daily Average Temperatures")
ax_main.set_xlabel("")
ax_main.set_ylabel("Average Daily Temperature C")

# --- Annual averages (horizontal lines) ---
annual_means = df_long.groupby(["variable", "year"])["value"].mean().reset_index()

for var in variables:
    for yr in annual_means["year"].unique():
        avg_val = annual_means.loc[
            (annual_means["variable"] == var) , "value"
        ]
        if not avg_val.empty:
            ax_main.axhline(y=avg_val.values[0],
                           color=palette[var], linestyle="--", linewidth=1.5,
                           alpha=0.7)

# --- Histograms (one per variable, stacked horizontally) ---
for var in variables:
    subset = df_long[df_long["variable"] == var]
    ax_hist.hist(subset["value"], bins=40, color=palette[var],
                 alpha=0.3, label=var, orientation="horizontal")

ax_hist.set_title("Value Distributions")
ax_hist.set_xlabel("Count")
ax_hist.set_ylabel("")  # share y with main

# Legend (only once)
ax_main.legend(title="")

plt.tight_layout()
plt.show()
```
